# 分词 / 词性标注 / 命名实体识别 / 成分句法分析

使用``HanLP``对产品名称、产品描述进行分词、词性标注、命名实体识别、成分句法分析。

```python
import json

products = json.load(open('products.json', 'r', encoding='utf-8'))
```

```python
import hanlp

HanLP = hanlp.load(hanlp.pretrained.mtl.CLOSE_TOK_POS_NER_SRL_DEP_SDP_CON_ELECTRA_BASE_ZH)

def tokenize(text):
    """
    对文本进行分词
    """
    doc = HanLP(text, tasks=['tok/fine', 'pos/ctb', 'ner/msra', 'con'], device="cuda")
    return doc
```

```python
from tqdm import tqdm

for i in tqdm(range(len(products))):
    product = products[i]
    product["tokenized_product_name"] = tokenize(product["product_name"])
    product["tokenized_sensory_characteristics"] = tokenize(product["sensory_characteristics"])
```

```python
# 写入到json文件
json.dump(products, open('tokenized_products.json', 'w', encoding='utf-8'), ensure_ascii=False, indent=4)
```
